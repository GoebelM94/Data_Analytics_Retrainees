```python
# neue Spalte für Alter

df['Alter'] = df['Teilnehmer_Adresse_Geburtsdatum']
df.head()
```

```python
# heutiges Datum
today = date.today()

# Funktion, die das Alter berechnet
def age(birthdate):
    
    # Boolean, der ausgibt ob heutiger Monat/Tag vor dem Geburtstags-Monat/Tag liegt
    one_or_zero = ((today.month, today.day) < (birthdate.month, birthdate.day))
    
    # Unterschied der Jahre zwischen heute und Geburtstag ausgeben
    diff = today.year-birthdate.year
    
    # der Unterschied in Jahren passt noch nicht genau, wenn der Monat/Tag des Geburtstags hinter dem heutigen Monat/Tag liegt
    # also muss das Ergebnis noch den one_or_zero Boolean berücksichtigen
    age = diff-one_or_zero
    
    return age

#print(age(date(2000,10,1)))
```

```python
# Spalte für Geburtsdatum in Datentyp datetime.date umwandeln
# und Alter in Spalte Alter berechnen

for i in range(0, df_len):
    df['Teilnehmer_Adresse_Geburtsdatum'].iloc[i] = datetime.date(df['Alter'].iloc[i])
    df['Alter'].iloc[i] = age(df['Teilnehmer_Adresse_Geburtsdatum'].iloc[i])
```

# Tests nach Abhängigkeit

**0. Tätigkeit_Status Unterscheidung (alle möglichen Status)**

```python
# weiteren DF erstellen ohne eigene Kategorien für die Tätigkeit Status 'geförderte Beschäftigung' und 'selbständige
# Tätigkeit' - diese werden in 'Sonstige' integriert

df_0 = df.copy()

for i in range(0, df_len):
    if df_0['Evaluation_TN_taetigkeit_status'].iloc[i] == 'geförderte Beschäftigung' or df_0['Evaluation_TN_taetigkeit_status'].iloc[i] == 'selbständige Tätigkeit':
        df_0['Evaluation_TN_taetigkeit_status'].iloc[i] = 'Sonstige'

df_0.head()
```

```python
# Kreuztabelle erstellen für Tätigkeit_Status x Alterskategorie

crosstab = pd.crosstab(df_0['Evaluation_TN_taetigkeit_status'],
                      df_0['Alterskategorie'],
                      margins=True, margins_name='Total')

crosstab
```

```python
# defining the table
data =[[12, 8, 2, 3],
      [73, 48, 12, 18],
      [34, 21, 7, 6],
      [69, 30, 12, 17],
      [42, 26, 7, 15]]

stat, p, dof, expected = stats.chi2_contingency(data)

# interpret p-value
alpha = 0.05
print('stat/chi-squared is ', stat)
print('p value is ', p)
print('dof is ', dof)
print('expected is ', expected)

if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

**_Ergebnis_**

```python
# Ergebnis
# alpha = 0,05; dof = 12 => Wert lt. Chi-Quadrat-Tabelle = 21,026
# Chi-Quadrat-Wert = 5,121
# 5,121 < 21,026 => H0 wird beibehalten
```

```python
# Berechnung Kontingenzkoeffizient

c = math.sqrt(stat / (df_len+stat))
print('C: ', c)
c_max = math.sqrt((min(5, 4)-1) / min(5, 4))
print('C_max: ', c_max)
c_norm = c/c_max
print('C_norm (normierter Kontingenzkoeffizient) ')
print('C_norm: ', c_norm)
```

**1. Tätigkeit_Status Unterscheidung 'versicherungspflichtige Beschäftigung' / 'x'**

```python
# defining the table
data_2 = [[69, 30, 12, 17], [161, 103, 28, 42]]
stat_2, p_2, dof_2, expected_2 = stats.chi2_contingency(data_2)

# interpret p-value
alpha = 0.05
print('stat/chi-squared is ' + str(stat_2))
print('p value is ' + str(p_2))
print('dof is ' + str(dof_2))
print(' expected is ' + str(expected_2))

if p_2 <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

**_Ergebnis_**

```python
# Ergebnis:
# alpha = 0,05; dof = 3 => Wert lt. Chi-Quadrat-Tabelle = 7,815
# Chi-Quadrat-Wert = 2,506
# 2,506 < 7,815 => H0 wird beibehalten
```

```python
# Berechnung Kontingenzkoeffizient

c_2 = math.sqrt(stat_2/(df_len+stat_2))
print('C: ', c_2)

c_max_2 = math.sqrt((min(4, 2)-1) / min(4,2))
print('C_max: ', c_max_2)

c_norm_2 = c_2/c_max_2
print('C_norm (normierter Kontingenzkoeffizient) prüft die Abhängigkeit zwischen zwei Variablen: 0 bedeutet komplette Unabhängigkeit, 1 bedeutet komplette Abhängigkeit der Variablen')
print('C_norm: ', c_norm_2)
```
